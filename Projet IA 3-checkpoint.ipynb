## Import  libraries

```python
import os
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score,
    classification_report, confusion_matrix, roc_auc_score,
    roc_curve, auc, precision_recall_curve
)

import joblib

# Setting visuels
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8,5)
RANDOM_STATE = 42
```

* ) In this cell, i imported all the necessary libraries for data manipulation, visualization, preprocessing, modeling, and evaluation.
* ) Seaborn and Matplotlib will be used for plotting, while scikit-learn and imbalanced-learn provide tools for model training and class balancing.

## Dataset loading

```python

data_file = Path.cwd() / "ai4i2020.csv"   

if not data_file.exists():
    raise FileNotFoundError(f"file not found : {data_file}\n"
                            "Put 'ai4i2020.csv' in the same folder as the Notebook.")
    
df = pd.read_csv(data_file)
print("Dataset loaded:", data_file)
print("Shape:", df.shape)
df.head(10)
```

* ) Here, i loaded the AI4I 2020 predictive maintenance dataset into a pandas DataFrame and inspect the first few rows to get an overview of the data and its structure.

## Exploration of the Dataset

```python

display(df.info())
display(df.describe().T)

print("Missing values per column:\n", df.isnull().sum())
print("\nDuplicated rows:", df.duplicated().sum())

# Remove duplicates (if any)
df = df.drop_duplicates().reset_index(drop=True)
print("Shape after deleting the duplicates:", df.shape)
```

* )  Checking the dataset information, descriptive statistics, and missing values.
* )  This helps to understand the type of each column, ranges, and if any preprocessing is needed 

## Verification of the distribution of Machine failure 

```python

if "Machine failure" in df.columns:
    sns.countplot(x="Machine failure", data=df)
    plt.title("Distribution — Machine failure")
    plt.show()
    print(df["Machine failure"].value_counts(normalize=True))
else:
    print("La colonne 'Machine failure' is absent.")
```

* ) Here i verified the data distribution and especially the machine failure values. 

##   Pairplotting and Correlation Matrix plotting

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
corr = df[num_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="RdBu_r")
plt.title(" correlation Matrix ")
plt.show()

# Pairplotting on a small subset
subset = ['Air temperature [K]', 'Process temperature [K]', 'Torque [Nm]', 'Tool wear [min]']
subset = [c for c in subset if c in df.columns]
if len(subset) >= 2:
    sns.pairplot(df[subset].sample(min(500, len(df))), diag_kind='kde') 
    plt.show()
```

## Feature Engineering

```python

required = ['Air temperature [K]', 'Process temperature [K]']
if all(c in df.columns for c in required):
    df['Delta_T'] = df['Process temperature [K]'] - df['Air temperature [K]']
    print("Feature 'Delta_T' .")
else:
    print("Colonnes nécessaires pour 'Delta_T' manquantes.")

# Features Selection
features = ['Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]',
            'Torque [Nm]', 'Tool wear [min]', 'Delta_T']
features = [f for f in features if f in df.columns]
print("Features retenues:", features)

# Target
if 'Machine failure' not in df.columns:
    raise KeyError("La colonne 'Machine failure' is mandatory.")
X = df[features].copy()
y = df['Machine failure'].copy()

# Final Verification 
X.head()
```

* ) By creating a new feature, Delta_T, representing the temperature difference. Feature engineering can improve model performance by adding meaningful information derived from existing features.

## Split train/test

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=RANDOM_STATE, stratify=y
)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
print("Distribution y_train:\n", y_train.value_counts(normalize=True))
```

* ) The dataset is splitted into training and test sets, maintaining the proportion of classes (stratified split).
* ) This ensures that models are trained and evaluated on separate data, avoiding overfitting.

## SMOTE + StandardScaler

```python
sm = SMOTE(random_state=RANDOM_STATE)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
print("After SMOTE — train shape:", X_train_res.shape)
print("Distribution after SMOTE:\n", y_train_res.value_counts(normalize=True))

scaler = StandardScaler()
X_train_res_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test)  
```

* ) SMOTE is used to oversample the minority class in the training set, balancing the dataset.
* ) Balanced data improves model performance, especially for metrics like recall and F1-score.
* ) Features are standardized to have zero mean and unit variance. This step ensures that models sensitive to feature scales (e.g., logistic regression) perform optimally.

## Utility Features: Training and Metrics functions

```python
def train_and_predict(model, X_train, y_train, X_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    return model, y_pred, y_prob

def print_metrics(name, y_true, y_pred, y_prob):
    print(f"=== {name} ===")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("F1-score:", f1_score(y_true, y_pred, zero_division=0))
    print("Precision:", precision_score(y_true, y_pred, zero_division=0))
    print("Recall:", recall_score(y_true, y_pred, zero_division=0))
    print("AUC-ROC:", roc_auc_score(y_true, y_prob))
    print("\nClassification Report:\n", classification_report(y_true, y_pred, zero_division=0))
```

## Models Training

```python
models = {}
# Logistic Regression
lr = LogisticRegression(max_iter=2000, random_state=RANDOM_STATE)
lr, y_pred_lr, y_prob_lr = train_and_predict(lr, X_train_res_scaled, y_train_res, X_test_scaled)
models['Logistic Regression'] = (lr, y_pred_lr, y_prob_lr)

# Random Forest
rf = RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE)
rf, y_pred_rf, y_prob_rf = train_and_predict(rf, X_train_res_scaled, y_train_res, X_test_scaled)
models['Random Forest'] = (rf, y_pred_rf, y_prob_rf)

# XGBoost
xgb = XGBClassifier(max_depth=5, learning_rate=0.1, n_estimators=200,
                    random_state=RANDOM_STATE, eval_metric='logloss', use_label_encoder=False)
xgb, y_pred_xgb, y_prob_xgb = train_and_predict(xgb, X_train_res_scaled, y_train_res, X_test_scaled)
models['XGBoost'] = (xgb, y_pred_xgb, y_prob_xgb)

print("Training Completed.")
```

I tried to train and use the different models like : linaear regression, random forest and XG boost
* )    A logistic regression model is trained on the resampled data. Also store predicted probabilities for later AUC-ROC evaluation.
* )  Random Forest is trained, providing a more complex model capable of capturing nonlinear relationships. Probabilities are stored for ROC analysis.   
* ) XGBoost is trained as a high-performance gradient boosting algorithm. It often outperforms other models on tabular datasets.

## Confusion Matrix and Models evaluation

```python
results = []
for name, (model, pred, prob) in models.items():
    print_metrics(name, y_test, pred, prob)
    results.append({
        'Model': name,
        'Accuracy': accuracy_score(y_test, pred),
        'F1': f1_score(y_test, pred, zero_division=0),
        'Precision': precision_score(y_test, pred, zero_division=0),
        'Recall': recall_score(y_test, pred, zero_division=0),
        'AUC': roc_auc_score(y_test, prob) })
    # Confusion matrix
    cm = confusion_matrix(y_test, pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix — {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
#Results
results_df = pd.DataFrame(results).sort_values(by='AUC', ascending=False)
display(results_df)
```

* ) Here ,I evaluated the different algorithms by calculating the accuracy, the F1-score, Recall , the AUC-ROC and the Precision.
* ) After calculating all the metrics , we had a report that includes the different results.
* )Then , by plotting the Confusion matrix i showed true vs predicted classes, and AUC-ROC indicates the model's ability to distinguish between classes.

## Roc-Curve and Precision -Recall Plotting

```python
plt.figure()
for name, (model, pred, prob) in models.items():
    fpr, tpr, _ = roc_curve(y_test, prob)
    plt.plot(fpr, tpr, label=f"{name} (AUC={auc(fpr,tpr):.3f})")
plt.plot([0,1],[0,1],'k--')
plt.title("ROC Curves")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()

# Precision-Recall
plt.figure()
for name, (model, pred, prob) in models.items():
    precision, recall, _ = precision_recall_curve(y_test, prob)
    plt.plot(recall, precision, label=name)
plt.title("Precision-Recall Curves")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend()
plt.show()
```

* ) ROC curves visualize model performance across thresholds. The closer the curve is to the top-left corner and the higher the AUC, the better the model distinguishes between failure and non-failure.
* ) Precision-Recall curves are important for imbalanced datasets. They show the trade-off between precision and recall.
* )  Higher area under the curve indicates a better model in correctly identifying failures.

## Feature importance (Random Forest)

```python
rf_model = models['Random Forest'][0]
importances = rf_model.feature_importances_
fi = pd.DataFrame({'feature': features, 'importance': importances}).sort_values('importance', ascending=False)
sns.barplot(x='importance', y='feature', data=fi)
plt.title("Feature importance — Random Forest")
plt.show()

print(fi)
```

  ## Saving of the Scaler and Definition of the best model 

```python
best_name = results_df.iloc[0]['Model']
best_model = models[best_name][0]

joblib.dump(scaler, "scaler.pkl")
joblib.dump(best_model, "best_model.pkl")
print(f"Saved scaler.pkl and best_model.pkl (best model: {best_name})")
```
